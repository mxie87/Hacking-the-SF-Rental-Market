```python
import pandas as pd
import numpy as np
```

```python
# load in dataframe
df = pd.read_csv('SFrentalsV2.csv')

# V2 was used because original csv had 1 error (2 sets of data in 1 row, moved 2nd set to bottom row )
```

```python
df.info()
# 3014 rows (including header) x 18 cols 
```

```python
df.head()
```

```python
# remove duplicates
df.drop_duplicates(keep=False, inplace=True)
```

```python
df.info()
# note that rows decreased from 3014 to 1006 (only 1005 unique rows)
```

```python
# Output CSV file with index, lat. and long. only. latittude and longitude used for reverse Geocoding
df['lat.'] = df['latitude,longitude'].str.strip('(').str.split(',').str[0].astype(float)
df['long.'] = df['latitude,longitude'].str.strip(')').str.split(',').str[1].astype(float)

df_geo=df.iloc[:,-2:]
df_geo.to_csv('reverse_geocoding.csv')
```

```python
# load in csv with zip codes assigned to each indexed row 
df_zipcodes = pd.read_csv('reverse_geocoding_geocodio_v2.csv')
df_zipcodes.head()
```

```python
df_zipcodes.set_index('Unnamed: 0',inplace=True)
df_zipcodes.head()
```

```python
# dropped 1 row where Zip is NaN
df_zipcodes = df_zipcodes[np.isfinite(df_zipcodes['Zip'])]
```

```python
df_zipcodes.info()
```

```python
# convert zipcodes from float to integer
df_zipcodes.Zip.astype(int)
```

```python
# merge dataframes
df_merged = pd.merge(df, df_zipcodes, left_index=True, right_index=True)
```

```python
df_merged.info()
```

```python
df_merged[df_merged['City']=='San Francisco'].count()
```

```python
# removed listings that were not located within San Francisco
# note that rows dropped from 1005 to 763
df_sf = df_merged[df_merged['City']=='San Francisco']
df_sf
```

```python
df_sf.to_csv('cleaned_sf.csv.csv')
```
